```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
df = pd.read_csv("IE KPMG_Demand Forecast dataset.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Iterate over unique ID_material values
for material_id in df['ID_material'].unique():
    # Filter the DataFrame for the specific ID_material
    material_df = df[df['ID_material'] == material_id]
    
    # Perform seasonal decomposition
    result = seasonal_decompose(material_df['Units'], model='additive', period=16)
    
    # Plot the decomposition
    result.plot()
    
    # Add a title with the ID_material name
    plt.title('Seasonal Decomposition for ID_material {}'.format(material_id))
    
    plt.show()
```
