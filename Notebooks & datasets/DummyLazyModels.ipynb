```python
import pandas as pd

def mean_absolute_percentage_error(y_true, y_pred):
    return (abs(y_true - y_pred) / y_true).mean() * 100
```

```python
df = pd.read_csv("monthly_data.csv")
```

```python
import pandas as pd

# Assuming your dataframe is named 'df'
mapping = {
    120004096: "Product1",
    120009814: "Product2",
    120009816: "Product3",
    120010342: "Product4",
    120010566: "Product5",
    120010970: "Product6",
    120011556: "Product7",
    120011782: "Product8",
    120012154: "Product9",
    120012606: "Product10",
    120014486: "Product11",
    120014488: "Product12",
    120015842: "Product13",
    120015996: "Product14"
}

df['ID_material'] = df['ID_material'].map(mapping)
```

# Global Mean

```python
# Iterate over each unique material in the DataFrame
for material in df['ID_material'].unique():
    # Filter the data for the current material
    material_data = df[df['ID_material'] == material]

    # Filter the data for each test set and create three separate dataframes for the current material
    test_set_1 = material_data[(material_data['Year'] == 2022) & (material_data['MonthName'].isin(['March', 'April']))]
    test_set_2 = material_data[(material_data['Year'] == 2022) & (material_data['MonthName'].isin(['June', 'July']))]
    test_set_3 = material_data[(material_data['Year'] == 2022) & (material_data['MonthName'].isin(['August', 'September']))]

    # Create the train set by excluding the rows used in the test sets for the current material
    train_set = material_data[~((material_data['Year'] == 2022) & (material_data['MonthName'].isin(['March', 'April', 'June', 'July', 'August', 'September'])))]

    # Append the rows of train_set and test_set_1 to create train_set_1
    train_set_1 = pd.concat([train_set, test_set_1])

    # Calculate the mean of "Units" column in the train set for the current material
    mean_unit = train_set['Units'].mean()
    mean_unit_t1 = train_set_1['Units'].mean()
    mean_unit_t2 = pd.concat([train_set_1, test_set_2])['Units'].mean()

    # Calculate the sum of "Units" column in each test set for the current material
    sum_1 = test_set_1['Units'].sum()
    sum_2 = test_set_2['Units'].sum()
    sum_3 = test_set_3['Units'].sum()

    # Append the results to the respective lists

    mape_1 = mean_absolute_percentage_error(sum_1, 2 * mean_unit)
    mape_2 = mean_absolute_percentage_error(sum_2, 2 * mean_unit_t1)
    mape_3 = mean_absolute_percentage_error(sum_3, 2 * mean_unit_t2)
    mean_mapes = (mape_1 + mape_2 + mape_3) / 3
    print(f"Mean MAPE for {material} is", mean_mapes.round(2))
```

```python
# Iterate over each unique material in the DataFrame
for material in df['ID_material'].unique():

    # Filter the data for the current material
    material_data = df[df['ID_material'] == material]

    # Filter the data for each test set and create three separate dataframes for the current material
    test_set_1 = material_data[(material_data['Year'] == 2022) & (material_data['MonthName'].isin(['March', 'April']))]
    test_set_2 = material_data[(material_data['Year'] == 2022) & (material_data['MonthName'].isin(['June', 'July']))]
    test_set_3 = material_data[(material_data['Year'] == 2022) & (material_data['MonthName'].isin(['August', 'September']))]

    # Create the train set by excluding the rows used in the test sets for the current material
    train_set = material_data[~((material_data['Year'] == 2022) & (material_data['MonthName'].isin(['March', 'April', 'June', 'July', 'August', 'September'])))]

    mean_units_by_month = train_set.groupby('MonthName')['Units'].median()
    mean_units_dict = mean_units_by_month.to_dict()

    # Append the rows of train_set and test_set_1 to create train_set_1
    train_set_1 = pd.concat([train_set, test_set_1])

    mean_units_by_month_1 = train_set_1.groupby('MonthName')['Units'].median()
    mean_units_dict_1 = mean_units_by_month_1.to_dict()

    train_set_2 = pd.concat([train_set, test_set_1])

    mean_units_by_month_2 = train_set_2.groupby('MonthName')['Units'].median()
    mean_units_dict_2 = mean_units_by_month_2.to_dict()


    # Calculate the mean of "Units" column in the train set for the current material
    mean_unit = (mean_units_dict["April"]+mean_units_dict["March"])/2
    mean_unit_t1 = (mean_units_dict_1["June"]+mean_units_dict_1["July"])/2
    mean_unit_t2 = (mean_units_dict_2["August"]+mean_units_dict_2["September"])/2

    # Calculate the sum of "Units" column in each test set for the current material
    sum_1 = test_set_1['Units'].sum()
    sum_2 = test_set_2['Units'].sum()
    sum_3 = test_set_3['Units'].sum()

    # Append the results to the respective lists

    mape_1 = mean_absolute_percentage_error(sum_1, 2 * mean_unit)
    mape_2 = mean_absolute_percentage_error(sum_2, 2 * mean_unit_t1)
    mape_3 = mean_absolute_percentage_error(sum_3, 2 * mean_unit_t2)
    mean_mapes = (mape_1 + mape_2 + mape_3) / 3
    print(f"Mean MAPE for {material} is", mean_mapes.round(2))
```

# Rolling Median Window

```python
import numpy as np

# Define the range of X values to try
x_values = range(1,36)

# Define a variable to store the best X value and the corresponding lowest mean MAPE
best_x = None
lowest_mape = np.inf

# Iterate over each X value
for X in x_values:
    mapes_trials = []  # Define the empty list to store MAPE results

    # Iterate over each unique material in the DataFrame
    for material in df['ID_material'].unique():
        # Filter the data for the current material
        material_data = df[df['ID_material'] == material]

        # Filter the data for each test set and create three separate dataframes for the current material
        test_set_1 = material_data[(material_data['Year'] == 2022) & (material_data['MonthName'].isin(['March', 'April']))]
        test_set_2 = material_data[(material_data['Year'] == 2022) & (material_data['MonthName'].isin(['June', 'July']))]
        test_set_3 = material_data[(material_data['Year'] == 2022) & (material_data['MonthName'].isin(['August', 'September']))]

        # Create the train set by excluding the rows used in the test sets for the current material
        train_set = material_data[~((material_data['Year'] == 2022) & (material_data['MonthName'].isin(['March', 'April', 'June', 'July', 'August', 'September'])))]

        # Calculate the median of "Units" column in the train set for the most recent X number of months
        train_set_X = train_set.tail(X)
        mean_unit_X = train_set_X['Units'].median()
        mean_unit_t1_X = pd.concat([train_set_X, test_set_1])['Units'].median()
        mean_unit_t2_X = pd.concat([train_set_X, test_set_1, test_set_2])['Units'].median()

        # Calculate the sum of "Units" column in each test set for the current material
        sum_1 = test_set_1['Units'].sum()
        sum_2 = test_set_2['Units'].sum()
        sum_3 = test_set_3['Units'].sum()

        # Append the results to the respective lists
        mape_1 = mean_absolute_percentage_error(sum_1, 2 * mean_unit_X)
        mape_2 = mean_absolute_percentage_error(sum_2, 2 * mean_unit_t1_X)
        mape_3 = mean_absolute_percentage_error(sum_3, 2 * mean_unit_t2_X)
        mean_mapes = (mape_1 + mape_2 + mape_3) / 3
        mapes_trials.append(mean_mapes)  # Store the mean MAPE in the list

    # Calculate the mean of mapes_trials for the current X value
    mean_mape = np.mean(mapes_trials)

    # Check if the current X value gives a lower mean MAPE than the previous lowest
    if mean_mape < lowest_mape:
        lowest_mape = mean_mape
        best_x = X

    print(f"Mean MAPE for X={X} is", mean_mape.round(2))

# Print the best X value and the corresponding lowest mean MAPE
print("Best X value:", best_x)
print("Lowest mean MAPE:", lowest_mape.round(2))
```
